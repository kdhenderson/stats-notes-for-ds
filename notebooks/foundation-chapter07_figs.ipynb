```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0
sigma = 1
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, mu, sigma)

# Critical values
alpha = 0.05
alpha_bonf = alpha / 2

z_alpha = norm.ppf(1 - alpha / 2)        # ≈ 1.96
z_bonf = norm.ppf(1 - alpha_bonf / 2)    # ≈ 2.24

# Plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Standard Normal", color="black")

# Shade unadjusted alpha regions
plt.fill_between(x, y, 0, where=(x <= -z_alpha) | (x >= z_alpha),
                 color='red', alpha=0.5, label="α = 0.05 (Unadjusted)")

# Shade Bonferroni adjusted alpha regions
plt.fill_between(x, y, 0, where=(x <= -z_bonf) | (x >= z_bonf),
                 color='blue', alpha=0.5, label="α' = 0.025 (Bonferroni)")

# Reference lines
plt.axvline(-z_alpha, color='red', linestyle='--', linewidth=1)
plt.axvline(z_alpha, color='red', linestyle='--', linewidth=1)
plt.axvline(-z_bonf, color='blue', linestyle='--', linewidth=1)
plt.axvline(z_bonf, color='blue', linestyle='--', linewidth=1)

# Labels and legend
plt.title("Bonferroni Correction and Type I Error Regions")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend(loc="upper left")
plt.grid(True, color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.tight_layout()
plt.show()
```
